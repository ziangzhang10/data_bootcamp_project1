```python
# Time for question 0 (just got this idea last-minute, might not use):
# Number of accidents vs. day
accident_df = accident_df.sort_values(["MONTH", "DAY"])

# Create groupby object to rank by accident by the day
accident_gby_day = accident_df[["MONTH", "DAY", "DAY_WEEK", "STATE"]].groupby(["MONTH", "DAY", "DAY_WEEK"])

accident_by_day = pd.DataFrame()
accident_by_day = accident_gby_day.count()
accident_by_day = accident_by_day.rename(columns = {"STATE": "Accidents"})
accident_by_day = accident_by_day.reset_index(drop=False, inplace=False)
accident_by_day["Day of year"] = accident_by_day.index.copy() + 1

# Pretty-display all rows
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(accident_by_day)
```

```python
# Plot instances of accidents vs day of year

x = accident_by_day["Day of year"]
y = accident_by_day["Accidents"]

color = ["steelblue"if item in ["Monday", "Tuesday", "Wednesday", "Thursday"] else "darksalmon"  for item in accident_by_day["DAY_WEEK"] ]

plt.xlabel('Day of the Year')
#plt.xticks(rotation=90)
plt.ylabel('Total Accidents in the US')

plt.scatter(x, y, marker=".", alpha=1, c=color)

plt.show()

# Save Figure
png_path = os.path.join("Charts", "total_accidents_by_day.png")
plt.savefig(png_path)
```

```python
# Plot specifically # accidents by day of the week 
accident_gby_day_week = accident_df[["DAY_WEEK", "STATE"]].groupby(["DAY_WEEK"])

accident_by_day_week = pd.DataFrame()
accident_by_day_week = accident_gby_day_week.count()
accident_by_day_week = accident_by_day_week.rename(columns = {"STATE": "Accidents"})
accident_by_day_week = accident_by_day_week.reset_index(drop=False, inplace=False)

# sort by day of the week
cat = CategoricalDtype(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ordered=True)
accident_by_day_week['DAY_WEEK'] = accident_by_day_week['DAY_WEEK'].astype(cat)
accident_by_day_week = accident_by_day_week.sort_values(["DAY_WEEK"])

# Pretty-display all rows
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(accident_by_day_week)
    
plt.xlabel('Weekday')
plt.xticks(rotation=90)
plt.ylabel('Total Accidents in the US')

color = ["steelblue"if item in ["Monday", "Tuesday", "Wednesday", "Thursday"] else "darksalmon"  for item in accident_by_day_week["DAY_WEEK"] ]

plt.bar(x = accident_by_day_week['DAY_WEEK'], height = accident_by_day_week["Accidents"], color = color, alpha = 1)

png_path = os.path.join("Charts", "total_accidents_by_week.png")
plt.savefig(png_path, bbox_inches='tight')
```

```python
# Plot specifically # accidents by the month
accident_gby_month = accident_df[["MONTH", "STATE"]].groupby(["MONTH"])

accident_by_month = pd.DataFrame()
accident_by_month = accident_gby_month.count()
accident_by_month = accident_by_month.rename(columns = {"STATE": "Accidents"})
accident_by_month = accident_by_month.reset_index(drop=False, inplace=False)

# sort by month
accident_by_month = accident_by_month.sort_values(["MONTH"])
accident_by_month["Month"] = ["January", "Feburary", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

# Pretty-display all rows
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(accident_by_month)
    
plt.xlabel('Month')
plt.xticks(rotation=90)
plt.ylabel('Total Accidents in the US')
    
plt.bar(x = accident_by_month['Month'], height = accident_by_month["Accidents"], color = 'steelblue', alpha = 1)

png_path = os.path.join("Charts", "total_accidents_by_month.png")
plt.savefig(png_path, bbox_inches='tight')
```

```python
# Time for the weather-per-day data
weather_day_houston = os.path.join("HistoricalWeatherByCity", "weather_day_houston.csv")
houston_df = pd.read_csv(weather_day_houston)

# Pretty-display all rows
# with pd.option_context('display.max_rows', None, 'display.max_columns', None):
#     display(houston_df)

weather_df = pd.DataFrame()
weather_df = houston_df.groupby("Weather").describe()
weather_df.columns = weather_df.columns.droplevel()
weather_df = weather_df["count"]
# weather_df = weather_df.rename(columns = {"count": "Days"})
weather_df = weather_df.reset_index(level = "Weather")
display(weather_df)

new_row2 = ["Rain", weather_df.loc[[4,5,7],["count"]].sum()[0]]
weather_df.loc[8] = new_row2

new_row1 = ["Fog/Smog/Smoke", weather_df.loc[[2,3,6],["count"]].sum()[0]]
weather_df.loc[9] = new_row1
weather_df = weather_df.loc[[0,1,8,9]]

weather_df
```

```python
# get the Houston-specific accident data
accident_df_houston = pd.DataFrame()

# code for Harris county is 201
accident_df_houston = accident_df.loc[(accident_df["STATE"] == "Texas")]
accident_df_houston = accident_df_houston.loc[accident_df["COUNTY"] == 201]
accident_df_houston = accident_df_houston[accident_df_houston["MONTH"] >= 6]
# Pretty-display all rows
# with pd.option_context('display.max_rows', None, 'display.max_columns', None):
#     display(accident_df_houston)
# accident_df_houston = accident_df_houston[["DAY", "MONTH", "YEAR"]]
# accident_df_houston["temp"] = "-"
# accident_df_houston["Date"] = accident_df_houston["YEAR"].astype(str) + accident_df_houston["temp"] + accident_df_houston["MONTH"].map("{:02}".format) + accident_df_houston["temp"] + accident_df_houston["DAY"].map("{:02}".format)

# create groupby object for Houston to get number of accidents per weather
accident_df_houston_gby = accident_df_houston[["YEAR", "WEATHER"]].groupby("WEATHER")
accident_weather_df = pd.DataFrame()
accident_weather_df = accident_df_houston_gby.describe()
accident_weather_df.columns = accident_weather_df.columns.droplevel()
accident_weather_df = accident_weather_df["count"]
# accident_weather_df = accident_weather_df.rename(columns = {"count": "Count"})
accident_weather_df = accident_weather_df.reset_index(level = "WEATHER")
accident_weather_df = accident_weather_df.rename(columns = {"WEATHER": "Weather"})
accident_weather_df = accident_weather_df.rename(columns = {"count": "Accidents"})

accident_weather_df.loc[[2], ["Weather"]] = "Fog/Smog/Smoke"
accident_weather_df.loc[[1], ["Weather"]] = "Clouds"
display(accident_weather_df)
```

```python
# merge the two datasets
merge_table = pd.merge(weather_df, accident_weather_df, on="Weather")
merge_table = merge_table.rename(columns = {"count": "Days"})
merge_table["Ratio of accident"] = merge_table["Accidents"]/merge_table["Days"]
merge_table
```

```python
# Build Pie Chart
plt.pie(merge_table["Days"], labels=merge_table["Weather"], autopct="%1.1f%%", shadow=True, startangle=0, 
        explode=[0.0, 0.0, 0.0, 0.0], colors = ["steelblue", "limegreen", "darksalmon", "gold"])
plt.title("% of weather by days")

# Save Figure
png_path = os.path.join("Charts", "weather_by_days.png")
plt.savefig(png_path, bbox_inches='tight')
```
