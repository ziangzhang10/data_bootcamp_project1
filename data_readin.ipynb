```python
# Time for question 0 (just got this idea last-minute, might not use):
# Number of accidents vs. day
accident_df = accident_df.sort_values(["MONTH", "DAY"])

# Create groupby object to rank by accident by the day
accident_gby_day = accident_df[["MONTH", "DAY", "DAY_WEEK", "STATE"]].groupby(["MONTH", "DAY", "DAY_WEEK"])

accident_by_day = pd.DataFrame()
accident_by_day = accident_gby_day.count()
accident_by_day = accident_by_day.rename(columns = {"STATE": "Accidents"})
accident_by_day = accident_by_day.reset_index(drop=False, inplace=False)
accident_by_day["Day of year"] = accident_by_day.index.copy() + 1

# Pretty-display all rows
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(accident_by_day)
```

```python
# Plot instances of accidents vs day of year

x = accident_by_day["Day of year"]
y = accident_by_day["Accidents"]

color = ["steelblue"if item in ["Monday", "Tuesday", "Wednesday", "Thursday"] else "darksalmon"  for item in accident_by_day["DAY_WEEK"] ]

plt.scatter(x, y, marker=".", alpha=1, c=color)

plt.show()
```

```python
# Plot specifically # accidents by day of the week 
accident_gby_day_week = accident_df[["DAY_WEEK", "STATE"]].groupby(["DAY_WEEK"])

accident_by_day_week = pd.DataFrame()
accident_by_day_week = accident_gby_day_week.count()
accident_by_day_week = accident_by_day_week.rename(columns = {"STATE": "Accidents"})
accident_by_day_week = accident_by_day_week.reset_index(drop=False, inplace=False)

# sort by day of the week
cat = CategoricalDtype(['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], ordered=True)
accident_by_day_week['DAY_WEEK'] = accident_by_day_week['DAY_WEEK'].astype(cat)
accident_by_day_week = accident_by_day_week.sort_values(["DAY_WEEK"])

# Pretty-display all rows
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(accident_by_day_week)
    
plt.bar(x = accident_by_day_week['DAY_WEEK'], height = accident_by_day_week["Accidents"], color = 'steelblue', alpha = 1)
```

```python
# Plot specifically # accidents by the month
accident_gby_month = accident_df[["MONTH", "STATE"]].groupby(["MONTH"])

accident_by_month = pd.DataFrame()
accident_by_month = accident_gby_month.count()
accident_by_month = accident_by_month.rename(columns = {"STATE": "Accidents"})
accident_by_month = accident_by_month.reset_index(drop=False, inplace=False)

# sort by month
accident_by_month = accident_by_month.sort_values(["MONTH"])

# Pretty-display all rows
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(accident_by_month)
    
plt.bar(x = accident_by_month['MONTH'], height = accident_by_month["Accidents"], color = 'steelblue', alpha = 1)
```

```python
# Get # of accidents vs weather
accident_gby_weather = accident_df[["WEATHER", "STATE"]].groupby(["WEATHER"])

accident_by_weather = pd.DataFrame()
accident_by_weather = accident_gby_weather.count()
accident_by_weather = accident_by_weather.rename(columns = {"STATE": "Accidents"})
accident_by_weather = accident_by_weather.reset_index(drop=False, inplace=False)

# sort by month
accident_by_weather = accident_by_weather.sort_values(["WEATHER"])

# Pretty-display all rows
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(accident_by_weather)
    
plt.bar(x = accident_by_weather['WEATHER'], height = accident_by_weather["Accidents"], color = 'steelblue', alpha = 1)
```
