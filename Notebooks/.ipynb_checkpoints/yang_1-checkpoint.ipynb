```python
csvpth = os.path.join("..", "Raw Data", "Drivers_by_States.csv")
driverdf = pd.read_csv(csvpth)
driver_clean_df = driverdf[["STATE","2015"]].dropna(axis = 0)
driver_clean_df.drop(labels = 51, inplace = True)
driver_clean_df
replace_dict = {"Alaska 2/" : "Alaska", "Hawaii 2/" : "Hawaii", "Dist. of Col." : "District of Colombia"}
driver_clean_df["STATE"].replace(replace_dict, inplace = True)
driver_clean_df
```
